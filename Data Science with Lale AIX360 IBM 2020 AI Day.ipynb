```python
#interpretable semi-automated pipeline
#lets pick between two different algorithms by automating their hyperparameter selection
custom_PCA = PCA.customize_schema(n_components = schemas.Int(min = 2, max = 30))

#simply by BOUNDING our decision space, we are able to change the way our pipeline behaves; comment out line below to see difference in model
KNN = KNN.customize_schema(n_neighbors=schemas.Int(min=1, max=10))
interpretable_pipeline = (Norm | NoOp | PCA)  >>  (KNN | DecisionTreeClassifier)
```

```python
#different color schems in the visualization indicate the state of that node in the pipeline
#white = trained, light blue = trainable, cyan = planned, purple = meta-model
interpretable_pipeline.visualize()
```

```python
#instantiate the trainable pipeline with the CASH of your choice (SMAC, GridSearch, etc.)
#cv = cross validation, 
ip_train =  Hyperopt(estimator = interpretable_pipeline, cv = 3, max_evals = 3)
```

```python
#can train the model now that it is in a trainable state
%time
ip_trained = ip_train.fit(train_X, train_y)
```

```python
#lets print out the pipeline
#Using a the normalized gaussian filter reduces the liklihood our x.values explode and become incomputable 
ip_trained.get_pipeline().visualize()
```

```python
#This is the summary of the iterations and the loss associated with each one
#this only works on the trained pipeline, will not work with the .auto_configure command
ip_trained.summary()
```

```python
best = ip_trained.summary().loss.argmin()
worst = ip_trained.summary().loss.argmax()
print(f'The best pipeline is {best} and the worst pipeline is {worst}')
```

```python
import sklearn.metrics

ip_y = ip_train.predict(test_X)
print(f'accuracy {sklearn.metrics.accuracy_score(test_y, ip_y):.1%}')
```
